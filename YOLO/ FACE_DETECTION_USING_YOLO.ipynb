```python
def imShow(path):
  import cv2
  import matplotlib.pyplot as plt

  img = cv2.imread(path)
  height, width = img.shape[0:2]
  resize_img=cv2.resize(img, (3*width, 3*height), interpolation = cv2.INTER_CUBIC)

  fig=plt.gcf()
  plt.imshow(cv2.cvtColor(resize_img,cv2.COLOR_BGR2RGB))
  plt.show()
```

 These are the new dimensions for the resized image. The

  image will be scaled up to three times its original width
  
  and height.

  plt.gcf(): Stands for "Get Current Figure". It returns
  
  the current figure object, which you can then modify or
  
  save.

interpolation = cv2.INTER_CUBIC: This specifies the

interpolation method. Cubic interpolation is used here,

which is a good choice for resizing images as it provides a

 good balance between speed and quality.



```python
imShow('/content/darknet/predictions.jpg')
```
