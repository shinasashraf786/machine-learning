Hiearchical clustering,also known as hiearchical cluster analysis(HCA)is an

unsupervised clustering algorithm that can be categorized in two ways:

agglomerative or divisive.Agglomerative clustering is considered a

"bottoms-up approach"Its data points are isolated as seperate groupings

initially,and they are merged together iteratively on the basis of similarity

until one cluster has been acheived.Four different methods are commonly used

to measure similarity:

In hiearchical clustering,linkages are used to calculate the distance between

clusters at each step of the agglomerative clustering process.Hiearchical

clustering builds a tree-like structure of clusters,where the leaves of the

tree represent individual data points,and the branches represent the merging

of cluster.(Dendogram)

Wards method minimizes the sum of squared differences within all clusters,

when merging two clusters,it chooses the pair that,when merged,results in the

smallest increase in the overall sum of squared differences

Average linkage:This method is defined by the mean distance between two points

in each cluster

Complete(or maximum)linkage.This method is defined by the

maximum distance between two points in each cluster

Single(or minimum)linkage:This method is defined by the minimum distance

between two points in each cluster these methods are used for getting the

distance b/w clusters
