```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.model_selection import GridSearchCV

from sklearn import metrics
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/machine learning/hyperparameter tuning/Copy of diabetes (1) (2) (1) (1).csv')
df.head()
```

```python
x = df.drop('Outcome',axis=1)
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train,y_test= train_test_split(x,y, test_size= 0.25, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
cls = KNeighborsClassifier(n_neighbors=5)
cls.fit(x_train,y_train)
y_pred = cls.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
help(cls)
```

```python
cls1 = KNeighborsClassifier()
```

**GridSearchCV**

```python
params = {'n_neighbors':[3,5,7,9,11], 'weights':['uniform','distance']}

clf = GridSearchCV(cls1,params,cv=10,scoring='accuracy')

clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
clf2 = KNeighborsClassifier(n_neighbors=9,weights='uniform')
clf2.fit(x_train,y_train)
y_pred = clf2.predict(x_test)
accuracy_score(y_test,y_pred)
```

**RandomSearchCV**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint as sp_randint
```

```python
clf = RandomForestClassifier(n_estimators=50)
```

```python
help(clf)
```

```python
param_dist = {"max_depth": [3,4,5,6,None],               #max depth - how much depth should decision tree comes -in what leafnode we should get output
              "max_features": sp_randint(1, 11),
              "min_samples_split": sp_randint(2,11),     #in what sample should we split
              "criterion": ["gini","entropy"]}
```

```python
randomCV = RandomizedSearchCV(clf, param_distributions=param_dist,cv=3)
```

```python
randomCV.fit(x_train,y_train)
print(randomCV.best_params_)
```

```python
print(randomCV.score(x_test,y_test))
```
